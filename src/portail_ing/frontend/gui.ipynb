```python
from ipyleaflet import Map, basemaps, WidgetControl, Circle
from ipywidgets import IntSlider, ColorPicker, Button, link, Text, HBox, VBox, Layout, FloatText
from panel.widgets import ArrayInput
from ipyleaflet import Map, Marker

center = (45.204793, -73.121558)

m = Map(center=center, zoom=5)

# Controls
circle_slider = IntSlider(description="Rayon (km):", min=0, max=250, value=50, layout=Layout(width="90%"))
button = Button(icon="dot-circle-o",  layout=Layout(width="12%", height='90%'), disabled=False)
lat = FloatText(placeholder="Lat", layout=Layout(width="40%"))
lon = FloatText(placeholder="Long", layout=Layout(width="40%"))

widget = VBox([HBox([button, lat, lon], width="200px"), circle_slider])
widget_control = WidgetControl(widget=widget, position="topright", max_width=300)
m.add(widget_control)

# Markers
# Pour un icon custom
#url = "http://127.0.0.1:8888/files/target.png"
#icon = Icon(icon_url=url, icon_size=[50, 50])
marker = Marker(location=center, draggable=True, visible=False)
circle = Circle(location=center, radius=50000, color="green", weight=3, visible=False)
link((circle_slider, "value"), (circle, "radius"), (lambda x: x * 1000, lambda x: x/1000))

m.add(marker)

log = {}

def create_marker_from_button(button):
    #button.disabled = not button.disabled
    if marker.visible:
        m.remove(circle)
    else:
        m.add(circle)
    marker.visible = not marker.visible
    
def create_marker_from_lat(event):
    log['lat'] = event
    lat = event["new"]
    marker.location = [lat, marker.location[1]]
    
def create_marker_from_lon(event):
    log['lon'] = event
    lon = event["new"]
    marker.location = [marker.location[0], lon]
    
# Links
button.on_click(create_marker_from_button)
lat.observe(create_marker_from_lat, "value")
lon.observe(create_marker_from_lon, "value")
link((marker, "location"), (circle, "location"))


def on_location_changed(event):
    log['changed'] = event
    location = event['new']
    # Need to unobserve to avoid infinite callback loops
    lon.unobserve(create_marker_from_lon, "value")
    lat.unobserve(create_marker_from_lat, "value")
    lon.value = location[1]
    lat.value = location[0]
    lat.observe(create_marker_from_lat, "value")
    lon.observe(create_marker_from_lon, "value")
    
marker.observe(on_location_changed, "location")
m
```
